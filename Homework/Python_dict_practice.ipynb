# Розділ 1. Словники (`dict`)

**Швидке нагадування:**

Словник — це колекція пар `ключ : значення`, створюється у фігурних дужках:
```python
student = {"name": "Олена", "age": 20, "city": "Київ"}
```

**Основні операції:**
- Доступ: `student["name"]`
- Додавання/зміна: `student["email"] = "olena@example.com"`
- Видалення: `del student["age"]`
- Перевірка: `if "name" in student:`
- Безпечне отримання: `student.get("phone")` — повертає `None`, якщо ключа немає
- Довжина: `len(student)` — кількість пар

**Перебір у циклі:**
- `for key in d.keys():` — перебір ключів
- `for value in d.values():` — перебір значень
- `for key, value in d.items():` — перебір пар

Дано словник з товарами магазину, де значення — це інший словник з ціною і кількістю на складі. Виконайте:
1. Виведіть загальну вартість всього товару на складі (ціна × кількість для кожного).
2. Виведіть назви товарів, яких залишилось менше 20 штук.
3. Виведіть найдорожчий товар.

````python
inventory = {
    "Ноутбук":   {"price": 18500, "stock": 12},
    "Телефон":   {"price": 9200,  "stock": 35},
    "Планшет":   {"price": 7800,  "stock": 8},
    "Навушники": {"price": 850,   "stock": 120},
    "Клавіатура":{"price": 1200,  "stock": 45},
    "Мишка":     {"price": 450,   "stock": 18}
}
# Ваш код:
# Виведіть загальну вартість всього товару на складі (ціна × кількість для кожного).

total_value = 0

for product, info in inventory.items():
  item_total = info["price"] * info["stock"]
  total_value += item_total
print(total_value)

# Виведіть назви товарів, яких залишилось менше 20 штук.

for product, info in inventory.items():
  if info["stock"] < 20:
    print(product)

# Виведіть найдорожчий товар.

max_price = 0
expensive_product = ""

for product, info in inventory.items():
  if info["price"] > max_price:
    max_price = info["price"]
    expensive_product = product
print(f"The most expensive product: {expensive_product}, price: {max_price} UAH")

````

> 💡 **Підказка:** Створіть порожній словник `revenue = {}`. У циклі для кожної покупки: якщо її категорія вже є у словнику — додайте ціну до існуючого значення, якщо немає — створіть запис з ціною як початковим значенням.

````python
purchases = [
    {"product": "Хліб",       "category": "Продукти",  "price": 25},
    {"product": "Молоко",     "category": "Продукти",  "price": 38},
    {"product": "Шампунь",    "category": "Гігієна",   "price": 120},
    {"product": "Зубна паста","category": "Гігієна",   "price": 65},
    {"product": "Сир",        "category": "Продукти",  "price": 180},
    {"product": "Книга",      "category": "Розваги",   "price": 220},
    {"product": "Мило",       "category": "Гігієна",   "price": 35},
    {"product": "Фільм",      "category": "Розваги",   "price": 150},
    {"product": "Ковбаса",    "category": "Продукти",  "price": 195}
]
# Ваш код:
revenue = {}
for item in purchases:
  cat = item["category"]
  pr = item["price"]
  if cat in revenue:
    revenue[cat] += pr
  else:
    revenue[cat] = pr
print(revenue)

````

2. Залежно від вибору — виконує відповідну дію
3. Завершується, коли користувач введе `вихід`

> 💡 **Підказка:** Використайте `while True:` (нескінченний цикл) і умову завершення з `break`, АБО змінну-прапорець, наприклад `running = True`.

````python
# Ваш код для завдання 22:
tel_book = {
    "Helen": "3134444444",
    "Alex":  "313488844444",
    "Natalie": "31349955544444",
    "Tom": "31344444000044",
    "Kim": "31344447777444"
}
running = True
while running != False:
  action = int(input("1 - пошук,\n2 - додати,\n3 - видалити,\n4 - вихід\n"))
  if action == 1:
    action2 = input("Name: ")
    if action2 in tel_book:
      print(tel_book[action2])
    else:
      print("Invalid contact")
  elif action == 4:
    running = False
    print("The program is over")
  elif action == 2:
    action3 = input("Enter name: ")
    action3_1 = input("Enter number: ")
    if action3 in tel_book:
      print("Such contact exists")
    else:
      tel_book[action3] = action3_1
      print("Contact added successfully")
    print(tel_book)
  elif action == 3:
    action4 = input("Enter name:")
    if action4 in tel_book:
      del tel_book[action4]
      print("Contact deleted")
      print(tel_book)
    else:
      print("Invalid contact")
  else:
    print("Incorrect number!")




````

### Завдання 45. Симулятор банкомату ⭐⭐⭐

Створіть змінну `balance = 5000`. У циклі `while` запитуйте у користувача дію:
- `баланс` — показати баланс
- `зняти` — запитати суму і зняти (якщо вистачає коштів)
- `поповнити` — запитати суму і додати до балансу
- `історія` — показати список усіх операцій
- `вихід` — завершити

Зберігайте кожну операцію у списку як словник `{"type": "зняти", "amount": 500}`.

> 💡 **Підказка:** Створіть `history = []` перед циклом. Кожну операцію додавайте через `.append()`.

````python
# Ваш код для завдання 45:
balance = 5000
history = []

while True:
    action = input("\nДія (баланс/зняти/поповнити/історія/вихід): ")

    if action == "баланс":
        print(f"Баланс: {balance} грн")

    elif action == "зняти":
        amount = int(input("Сума для зняття: "))
        if amount > balance:
            print("Недостатньо коштів!")
        else:
            balance -= amount
            history.append({"type": "зняти", "amount": amount})
            print(f"Знято {amount} грн. Баланс: {balance} грн")

    elif action == "поповнити":
        amount = int(input("Сума для поповнення: "))
        balance += amount
        history.append({"type": "поповнити", "amount": amount})
        print(f"Поповнено {amount} грн. Баланс: {balance} грн")

    elif action == "історія":
        if not history:
            print("Історія порожня")
        else:
            for op in history:
                print(f"{op['type']}: {op['amount']} грн")

    elif action == "вихід":
        print("До побачення!")
        break

    else:
        print("Невідома дія!")
````

### Завдання 46. Менеджер задач (todo) ⭐⭐⭐

Створіть програму-менеджер задач. Задачі зберігаються у списку словників: `{"task": "...", "done": False}`. У циклі `while` запитуйте дію:
- `додати` — додати нову задачу
- `показати` — показати всі задачі з номерами і статусом
- `виконати` — за номером позначити задачу як виконану
- `видалити` — за номером видалити задачу
- `вихід` — завершити

````python
# Ваш код для завдання 46:
tasks = []

while True:
    action = input("\nДія (додати/показати/виконати/видалити/вихід): ")

    if action == "додати":
        task = input("Назва задачі: ")
        tasks.append({"task": task, "done": False})
        print(f"Задачу '{task}' додано!")

    elif action == "показати":
        if not tasks:
            print("Список задач порожній")
        else:
            for i, t in enumerate(tasks):
                status = "✅" if t["done"] else "❌"
                print(f"{i + 1}. {status} {t['task']}")

    elif action == "виконати":
        num = int(input("Номер задачі: ")) - 1
        if 0 <= num < len(tasks):
            tasks[num]["done"] = True
            print(f"Задачу '{tasks[num]['task']}' виконано!")
        else:
            print("Невірний номер!")

    elif action == "видалити":
        num = int(input("Номер задачі: ")) - 1
        if 0 <= num < len(tasks):
            removed = tasks.pop(num)
            print(f"Задачу '{removed['task']}' видалено!")
        else:
            print("Невірний номер!")

    elif action == "вихід":
        print("До побачення!")
        break

    else:
        print("Невідома дія!")
````

Реалізуйте гру: дано стартове слово. Гравець має ввести наступне слово, яке починається на ОСТАННЮ літеру попереднього. Якщо слово вже використовувалось — програш. Якщо слово починається не на ту літеру — програш. Цикл триває, поки гравець не програє або не введе `здаюсь`.

> 💡 **Підказка:** Зберігайте всі використані слова у списку. Першу і останню літеру отримуйте через індекси: `word[0]` і `word[-1]`.

````python
# Ваш код для завдання 52:
start_word = "слово"
used_words = [start_word]
current_word = start_word

print(f"Стартове слово: {current_word}")

while True:
    next_letter = current_word[-1]
    user_word = input(f"Введіть слово на літеру '{next_letter}': ")

    if user_word == "здаюсь":
        print(f"Ви здались! Використано слів: {len(used_words)}")
        break

    if user_word[0] != next_letter:
        print(f"Програш! Слово має починатись на '{next_letter}'")
        break

    if user_word in used_words:
        print(f"Програш! Слово '{user_word}' вже використовувалось")
        break

    used_words.append(user_word)
    current_word = user_word
    print(f"Добре! Використано слів: {len(used_words)}")

print(f"Всі слова: {used_words}")
````

### Завдання 53. Аналіз журналу ⭐⭐⭐

Дано список словників-записів журналу логів. Кожен запис: `{"level": "INFO"|"WARNING"|"ERROR", "message": "...", "hour": ...}`. Створіть звіт:
1. Скільки записів кожного рівня (словник)
2. У яку годину було найбільше ERROR-ів
3. Усі повідомлення з рівнем ERROR
4. Чи були години без жодного ERROR (з тих, у яких є записи)

````python
logs = [
    {"level": "INFO",    "message": "Сервер запущено",     "hour": 8},
    {"level": "INFO",    "message": "Користувач увійшов", "hour": 9},
    {"level": "WARNING", "message": "Повільний запит",     "hour": 9},
    {"level": "ERROR",   "message": "Помилка БД",          "hour": 10},
    {"level": "ERROR",   "message": "Тайм-аут запиту",     "hour": 10},
    {"level": "INFO",    "message": "Запит виконано",      "hour": 11},
    {"level": "WARNING", "message": "Низько місця на диску","hour": 12},
    {"level": "ERROR",   "message": "Помилка БД",          "hour": 14},
    {"level": "INFO",    "message": "Резервна копія",      "hour": 15},
    {"level": "ERROR",   "message": "Збій авторизації",    "hour": 14},
    {"level": "ERROR",   "message": "Помилка БД",          "hour": 14},
    {"level": "INFO",    "message": "Користувач вийшов",   "hour": 17}
]
# Ваш код:
# 1. Кількість записів кожного рівня
by_level = {}
for log in logs:
    l = log['level']
    if l not in by_level:
        by_level[l] = 0
    by_level[l] += 1
print(f"По рівнях: {by_level}")

# 2. Година з найбільшою кількістю ERROR
errors_by_hour = {}
for log in logs:
    if log['level'] == "ERROR":
        h = log['hour']
        if h not in errors_by_hour:
            errors_by_hour[h] = 0
        errors_by_hour[h] += 1
worst_hour = max(errors_by_hour, key=errors_by_hour.get)
print(f"\nНайбільше ERROR о {worst_hour}:00 ({errors_by_hour[worst_hour]} помилок)")

# 3. Всі ERROR повідомлення
print("\nERROR повідомлення:")
for log in logs:
    if log['level'] == "ERROR":
        print(f"  [{log['hour']}:00] {log['message']}")

# 4. Години без ERROR
all_hours = set()
error_hours = set()
for log in logs:
    all_hours.add(log['hour'])
    if log['level'] == "ERROR":
        error_hours.add(log['hour'])
safe_hours = all_hours - error_hours
print(f"\nГодини без ERROR: {sorted(safe_hours)}")
````

Дано великий датасет — список замовлень. Кожне замовлення: словник з полями `customer`, `city`, `product`, `category`, `price`, `quantity`, `month`.
