````python
teams = ["Динамо", "Шахтар", "Зоря", "Дніпро", "Ворскла", "Олександрія"]
points = [64, 67, 52, 48, 39, 45]
# Ваш код:
place = 1
while len(points) > 0:
  i = points.index(max(points))
  print(f"{place}. {teams[i]} — {points[i]} очок")
  teams.pop(i)
  points.pop(i)
  place += 1
````
